### Let's say that we have a line whose equation is y = -0.6x + 4. For the point (x,y) = (-5, 3), apply the absolute trick to get the new equation for the line, using a learning rate of alpha=0.1alpha = 0.1alpha=0.1.

```python
import matplotlib.pyplot as plt
import numpy as np

scale_factor = 10

plt.xlim([-10,10])
plt.ylim([-10,10])

x=np.linspace(-10,10,100)
y = -0.6*x + 4
p, q = -5, 3
q1 = -0.6*p + 4
m = -0.6
yint = 4
alpha = 0.1

plt.plot(x, y)
plt.plot(-5,3, 'bo')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

### Absolute Trick

```python
def absoluteTrick(x, y, p, q, q1, alpha, m, yint):
    if (q-q1) > 0:
        y1 = (m + p*alpha)*x + (yint + alpha)
    else:
        y1 = (m - p*alpha)*x + (yint - alpha)
        
    plt.xlim([-10,10])
    plt.ylim([-10,10])
    plt.plot(x, y1, '-r', label='y=-.6x+4')
    plt.yticks()
    plt.plot(-5,3, 'bo')
    
    plt.grid()
    plt.show()
```

```python
absoluteTrick(x, y, p, q,q1,alpha, m, yint)
```

## Squared Trick

```python
def squareTrick(x, y, p, q, q1, alpha, m, yint):
    y1 = (m + p*alpha*(q-q1))*x + (yint + alpha*(q-q1))
    plt.xlim([-10,10])
    plt.ylim([-10,10])
    plt.plot(x, y1, '-r', label='y=-.6x+4')
    plt.plot(-5,3, 'bo')
    plt.grid()
    plt.show()
```

```python
squareTrick(x, y, p, q, q1, alpha, m, yint)
```
